```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([33,45,32,69,89,78,56])
plt.pie(y)
plt.show()
```

```python
#adding the labels to the pie chart 
import matplotlib.pyplot as plt 
import numpy as np
y=np.array([34,45,67,78])
mylabels=["mango","apple","grapes","pineaaple"]
plt.pie(y,labels=mylabels)
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np
y=np.array([34,45,67,78])
mylabels=["mango","apple","grapes","pineaaple"]
plt.pie(y,labels=mylabels,shadow=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "b", "#4CAF50"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend(title="Four Fruits")
plt.show()
```
