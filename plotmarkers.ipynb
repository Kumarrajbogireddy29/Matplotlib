```python
you can use the keyword argument marker to emphasize the each point 
with a specified marker 
```

```python
import matplotlib.pyplot as plt
import numpy as np
ypoints= np.array([1,3,5,8])
plt.plot(ypoints,marker='o')
plt.show()
```

```python
# Format strings
 youc an use the shortcut string notation parameter to specify the 
marker This parameter is also called as fmt is written with syntax
marker|line|color
```

```python
import matplotlib.pyplot as plt
import numpy as np
ypoints=np.array([3,7,9,0])
plt.plot(ypoints,'o:r')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='o',ms=20)
plt.show()
```

```python
#marker color
you can use the kyword argument  markeredgecolor or the shortest mec to set the color of teh edge of the a
markers
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r')
plt.show()
```

```python
#you can use the keyword argument markerfacecolor or the shorest 
mfc to set the color inside the edge of the markers
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = '*', ms = 20, mec = 'r')
plt.show()
```

```python
#you can use the mec and mfc arguments to color the entire marker 
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r',mfc='r')
plt.show()
```
